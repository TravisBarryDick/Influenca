Construct the counter example and do some other initialization

```julia
using Influenca, PyPlot
g = Graph(4) 
add_edge!(g, 1, 2)
add_edge!(g, 2, 3)
add_edge!(g, 3, 1)
Î²s = linspace(0, 1, 50);
```

# Checking Immunization Calculations

```julia
plot(Î²s, [evaluate_immunization(g, Î², [1,2], m=2) for Î² in Î²s], "b-", label="optimal simulated")
plot(Î²s, [1.0 for Î² in Î²s], "b.", label="optimal predicted")
plot(Î²s, [evaluate_immunization(g, Î², [1,4], m=2) for Î² in Î²s], "r-", label="sub-optimal simulated")
plot(Î²s, [1 + 2Î²/3 for Î² in Î²s], "r.", label="sub-optimal predicted")
xlabel("\$\\beta\$")
ylabel("Average number of infections")
title("Immunization")
legend(loc="upper left");
```

```julia
plot(Î²s, [evaluate_influence(g, Î², [1,4]) for Î² in Î²s], "b-", label="optimal simulated")
plot(Î²s, [2*(1+Î²+Î²^2-Î²^3) for Î² in Î²s], "b.", label="optimal predicted")
plot(Î²s, [evaluate_influence(g, Î², [1,2]) for Î² in Î²s], "r-", label="sub-optimal simulated")
plot(Î²s, [3 - (1-Î²)^2 for Î² in Î²s], "r.", label="sub-optimal predicted")
xlabel("\$\\beta\$")
ylabel("Average number of infections")
title("Influence Maximization")
legend(loc="upper left");
```
